```python
from tabulate import tabulate
p2p_inv_df = pd.read_sql('select * from p2p_invest',engine)
print tabulate(p2p_inv_df.describe(),headers = 'keys', tablefmt='psql')
p2p_inv_df.head()
#p2p_inv_df = pd.read_csv('p2p_inv_df.csv',encoding = 'utf-8') 을 실행 시켜서 AWS서버와 MY-SQL서비스 없이 실행 할 수 있음
```

* 핀다 사이트에는 p2p투자 상품에 대해서 사용자들의 리뷰가 각각의 상품이 아닌 회사별로 나타나므로 각각의 상품에 대한 사용자들의 반응을 알아보기 위해서는 기업별 리뷰 수를 파악할 필요가 있다. 
* 피플펀드의 제품이 총 21개의 리뷰를 받으며 가장 많은 리뷰를 받고 있고 총 133개의 상품이 핀다에 게재되었다. 
* 투게더의 경우 가장 많은 269개의 제품이 게재되어있지만 단 하나의 리뷰에 그치고 있다. 

```python
import re
def period_conv(strs):
    return int(re.search('\d+',strs).group())
def grade_conv(strs):
    try:
        return int(re.search("\d*\.?\d+",strs).group())

    except:
        return 0
p2p_inv_df['period'] = p2p_inv_df['period'].apply(grade_conv)
p2p_inv_df['grade'] = p2p_inv_df['grade'].apply(grade_conv)

p2p_inv_df.groupby('company').aggregate({'review_counts': np.mean,'average_rate': np.mean, 'period':np.mean,'default_rate':np.mean,
                                         'productname':pd.Series.nunique}).sort_values('review_counts',ascending = False)
```

* 모델을 통해 부도율, 투자기간, 투자 상품의 등급, 등을 고려하였을 때 가장 높은 이자율을 지급 하는 상품이 무엇인지 찾아 본다. 
* Linear Regression 모델과 Random Forest Regressor모델 중 더 높은 성능을 나타내는 Random Forest Regressor모델을 p2p투자 상품의 적정 이자율을 예측하기 위한 최종 모델로 선정 
* 데이터의 전처리는 pipeline을 통해서 실행 

```python
from sklearn.ensemble import ExtraTreesClassifier,RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler,OneHotEncoder,LabelEncoder
from sklearn.pipeline import make_pipeline,make_union
from sklearn.cross_validation import cross_val_score

# p2p투자상품의 투자 등급과 , 발행 회사의 데이터 전처리를 위한 class
class str_extractor(object):
    def __init__(self):
        pass
    def transform(self,df):
        lst = ['grade','company']
        le = LabelEncoder()
        return df[lst].apply(le.fit_transform)

    
    def fit(self,df,y = None):
        return self
    
class float_extractor(object):
    
    def __init__(self):
        pass
    
    def transform(self,df):
        lst = ['default_rate','period','target']
        return df[lst]
    def fit(self,df,y = None):
        return self    
    
class DenseTransformer(object):

    def transform(self, X, y=None, **fit_params):
        return X.todense()

    def fit_transform(self, X, y=None, **fit_params):
        self.fit(X, y, **fit_params)
        return self.transform(X)

    def fit(self, X, y=None, **fit_params):
        return self
    

pl1 = make_pipeline(float_extractor(),StandardScaler())
pl2 = make_pipeline(str_extractor(),OneHotEncoder(),DenseTransformer())
pl3 = make_union(pl1,pl2)
final_features = pl3.fit_transform(p2p_inv_df)
linearmodel = LinearRegression()
forest = RandomForestRegressor(n_estimators=450, random_state=0)
cross_score =-np.mean(cross_val_score(forest,final_features,p2p_inv_df['interest'],'neg_mean_squared_error',cv=5))
linear_cross_score =-np.mean(cross_val_score(linearmodel,final_features,p2p_inv_df['interest'],'neg_mean_squared_error',cv=5))
print 'Random Forest Regressor의 mean squared error',cross_score
print 'Linear Regression 모델의 mean squared error',float(linear_cross_score)
```

* 모델로 예측한 값에 비해서 높은 이자율을 지급하는 상품을 바탕으로 높은 수익률을 제공하는 상품 검색
    * 실제로 (실제금리 - 모델 예측값)이 가장 큰 상품을 검색 하였을 때 해당 상품이 속한 보유하고 있는 상품 등급 투자기간, 부도율 등을 고려해 보았을때 해당 등급에 비해 높은 이자율을 제공하는 상품이라는 것을  확인할 수 있었다. ->기존 핀다 사이트에서 제공해 주지 않았던 p2p 투자 상품에 대한 새로운 투자 기준 제시 

```python
print'4등급 투자상품 평균금리 :', p2p_inv_df.groupby('grade').mean()['interest'].ix[4]
print'24개월 투자상품 평균 금리:',p2p_inv_df.groupby('period').mean()['interest'].ix[24]
model =forest.fit(final_features,p2p_inv_df['interest'])
sns.regplot(model.predict(final_features),(p2p_inv_df['interest']))
plt.xlabel(u'모델로 측정한 예측 투자 이자율값')
plt.ylabel(u'실제 이자율')
p2p_inv_df['costeffect'] =model.predict(final_features)- p2p_inv_df['interest'] # 모델예측값 - 실제값
p2p_inv_df.sort_values('costeffect')[['productname','company','grade','interest','period','default_rate','target']][:10]
```
